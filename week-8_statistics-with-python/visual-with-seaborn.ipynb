```python
import seaborn as sns
import pandas as pd
import statsmodels.stats.diagnostic as stats

# For reference on visualization with Seaborn, go to
# https://seaborn.pydata.org/api.html#distribution-api
```

```python
tips = pd.read_csv('./tips.csv')
penguins = pd.read_csv('./penguins.csv')
```

### Relations

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

### Distributions

```python
penguins.head()
```

```python
sns.displot(data=penguins, x='flipper_length_mm')
```

```python
sns.displot(data=penguins, x='flipper_length_mm', kde=True)
```

```python
sns.displot(data=penguins, x='flipper_length_mm', col='sex', hue='species', kind='kde')
```

### Categories

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```
